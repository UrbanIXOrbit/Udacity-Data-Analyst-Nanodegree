### A report of the insight and visualization of the analysis of the we rate dogs datasets.

Dogs happen to be the most common household pet. Dog lovers have found the we rate dogs twitter handle to a forum for sharing pictures of dogs and rating them.
An analysis of the we rate dog twitter dataset was conducted and it shows a couple of things about users engagement and about the breeds of dogs users love most. 

The following insights were drawn from the analysis of the datasets. The first two insight has both public for the users of the we rate dogs and internal relevance for the we rate dog organization, the third insight is specifically for the internal use of the we rate dog organization.
